```python
import pandas as pd 
```

```python
dib=pd.read_csv("Diabetes_CDDS_dataset.csv") #we are using diabetes dataset to predict diabetes using machine learning algorithm
```

```python
dib # in this data set all the patients are womens
```

```python
dib.columns #this shows all the the coumns
```

```python
dib.count() 
```

```python
from matplotlib import pyplot as plt
plt.plot(dib["Age"],dib["Outcome"])
plt.xlabel( "Age")
plt.ylabel("Outcome/diabetic")
plt.title("diabetic Vs Age")
plt.show()
```

```python
X=dib.drop(columns="Outcome").values  # this took all the columns other than Outcome column 
#we can also write X= dib.iloc[:, [2,3]].values 2 and 3 represent the independent variable 
y=dib.Outcome.values#this takes only Output cloumn there 
# we can also write it as y= dib.iloc[:, 8].values  9 represent that Output column is at 8th index
```

```python
from sklearn.model_selection import train_test_split #here we are spilitting the data for training and testing 
X_train , X_test ,y_train ,y_test= train_test_split(X,y,test_size=0.20,random_state=0) # y is basically the part which the model will pridect
```

```python
print("X_train ",X_train.shape  )
print(" X_test ", X_test.shape )
print("Y_train ",y_train.shape  )
print("Y_test ",y_test.shape )
print("X",X.shape)
print("Y",y.shape)
```

```python
 from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)

X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

lor = LogisticRegression(solver="liblinear")  # optimal for binary + small dataset
lor.fit(X_train, y_train)
y_pred_lor = lor.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
acc1=accuracy_score(y_test,y_pred_lor)
acc1
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier(random_state=100 , max_depth=10, max_features= None, min_samples_leaf=15)
```

```python
dtc.fit(X_train, y_train)
y_pred_dtc= dtc.predict(X_test)
```

```python
acc2=accuracy_score(y_test,y_pred_dtc)
acc2
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc= RandomForestClassifier( random_state=100,  max_features= None, min_samples_leaf=15,oob_score=True, n_estimators=100)
rfc.fit(X_train ,y_train)
y_pred_rfc= rfc.predict(X_test)
```

```python
acc3=accuracy_score(y_test,y_pred_rfc)
acc3
```

```python
from sklearn.neighbors import KNeighborsClassifier
knc=KNeighborsClassifier(n_neighbors=10)
knc.fit(X_train,y_train)
y_pred_knc= knc.predict(X_test)
```

```python
acc4=accuracy_score(y_test,y_pred_knc)
acc4
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)
y_pred_nb = nb_model.predict(X_test)

acc5 = accuracy_score(y_test,y_pred_nb)
print (acc5)
```

```python
from xgboost import XGBClassifier


model = XGBClassifier(n_estimators=300, learning_rate=0.05, max_depth=6, eval_metric='logloss')


model.fit(X_train, y_train)


y_pred_xgb = model.predict(X_test)

# Accuracy
acc6 = accuracy_score(y_test, y_pred_xgb)
print("XGBoost Accuracy:", acc6)
```

```python
from sklearn.svm import SVC

svm_model = SVC(kernel='linear')   


svm_model.fit(X_train, y_train)


y_pred_svm= svm_model.predict(X_test)


acc7 = accuracy_score(y_test, y_pred_svm)
print("SVM Accuracy:", acc7)
```

```python
print(" The predicted accuracy for logistic Regression  :",acc1  )
print(" The predicted accuracy for decision tree  :",acc2 )
print(" The predicted accuracy for  random forest  :",acc3 )
print(" The predicted accuracy for k neighbours  :",acc4 )
print(" The predicted accuracy for naive bayes :",acc5 )
print("The predicted accuracy for XGBoost Accuracy:", acc6)
print("The predicted accuracy for SVM Accuracy:", acc7)
```
